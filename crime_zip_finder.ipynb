```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json
from api_keys import g_key
```

```python
csv_path = "Resources/crime_data_parsed.csv"
crime_data = pd.read_csv(csv_path)
                
```

```python
addresses = crime_data["INCIDENT_ADDRESS"].tolist()
```

```python
crime_zip =[]

#set the api key
params = {
    "key": g_key}
#create a for loop to use iterrows() to loop through the crimes in the dataframe and make an api request for each lat/long
#to return the zip

# for index, row in crime_data.iterrows():

for a in addresses:

        print (a)
#         loc = row["INCIDENT_ADDRESS"]
        loc = a
#         #update the location parameter to be the address of the crime's lat/lng
#         params["latlng"] = f"{lat},{lng}"
        params["address"]=loc
        #update the url to include the parameters
        base_url = "https://maps.googleapis.com/maps/api/geocode/json?"
        response = requests.get(base_url, params=params)
        #get a json response
        crime_response = response.json()
#         print(json.dumps(crime_response, indent=4, sort_keys=True))
        #print(crime_response)
        #crime_response=crime_response["results"]
       # try to append the data from the response
        try:
            crime_zip.append(crime_response['results'][0]["address_components"][7]["long_name"])

        except:
            crime_zip.append("N/A")
```

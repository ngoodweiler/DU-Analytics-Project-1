```python
response
```

```python
listings = response["listings"]
listings
```

```python
addresscount=-1
addresses = []
for address in listings:
    addresscount = addresscount+1
    address = listings[addresscount]["address"]
    addresses.append(address)
addresses
```

```python
sold_datecounter=-1
sold_dates = []
for sold_date in listings:
    sold_date = sold_datecounter+1
    sold_date = listings[sold_datecounter]["sold_date"]
    sold_dates.append(sold_date)
sold_dates
```

```python
listings = response["listings"]
listings
```

```python
counter=-1
addresses = []
sold_dates = []
prices = []
zips = []
lats = []
lons = []
for listing in listings:
    counter = counter+1
    address = listings[counter]["address"]
    sold_date = listings[counter]["sold_date"]
    price = listings[counter]["price"]
    lat = listings[counter]["lat"]
    lon = listings[counter]["lon"]
    zip = int(address[-5:])
    addresses.append(address)
    sold_dates.append(sold_date)
    prices.append(price)
    lats.append(lat)
    lons.append(lon)
    zips.append(zip)
    
```

```python
re_df = pd.DataFrame()
re_df["Address"]=addresses
re_df["Sold Date"]=sold_dates
re_df["Prices"]=prices
re_df["Lat"]=lats
re_df["Lon"]=lons
re_df["Zip"]=zips
re_df
```
